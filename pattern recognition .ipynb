```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Africa Food Production (2004 - 2013).csv')
df.head()
```

```python
df1=pd.read_csv('Africa Food Supply (2004 - 2013).csv')
df1.head()
```

```python
oat=df[df['Item']=='Oats']
oat.groupby('Country')['Value'].sum().sort_values(ascending=False)
```

```python
#trend in production
sns.lineplot(df.Year,df.Value)
```

```python
#trend in supply
sns.lineplot(df1.Year,df1.Value)
```

```python
sns.boxplot(data=df)
```

```python
df2=df1.drop('Year',axis='columns')
d3=pd.concat([df,df2],axis=1)
pd.pivot_table(d3, index='Year', values=['Value'])
```
